```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load
city_weather_data = "data/cities.csv"

# Read School and Student Data File and store into Pandas Data Frames
city_weather_df = pd.read_csv(city_weather_data)

city_weather_df.head()
```

```python
city_weather_df = pd.DataFrame.set_index(city_weather_df,'Record No.')
city_weather_df
```

```python
# https://stackoverflow.com/questions/50807744/apply-css-class-to-pandas-dataframe-using-to-html/50939211
# Pandas' to_html simply outputs a large string containing HTML table markup. 
# The classes argument is a convenience handler to give the <table> a class 
# attribute that will be referenced in a previously created CSS document that 
# styles it. Therefore, incorporate to_html into a wider HTML document build 
# that references an external CSS.
# Interestingly, to_html adds dual classes <table class="dataframe mystyle"> 
# which can be referenced in CSS individually, .dataframe {...} .mystyle{...}, 
# or together .dataframe.mystyle {...}

pd.set_option('colheader_justify', 'center')   # FOR TABLE <th>

html_string = '''
<html>
  <head><title>HTML Pandas Dataframe with CSS</title></head>
  <link rel="stylesheet" type="text/css" href="df_style.css"/>
  <body>
    {table}
  </body>
</html>.
'''

# OUTPUT AN HTML FILE
with open('data/city_weather.html', 'w') as f:
    f.write(html_string.format(table=city_weather_df.to_html(classes='mystyle')))
```
